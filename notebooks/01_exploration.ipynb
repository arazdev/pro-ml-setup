## 1. Install and Import Required Libraries

```python
import sys
import os
from pathlib import Path

# Add project root to path so we can import src modules
PROJECT_ROOT = Path(os.getcwd()).parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Import our project modules
from src.data_loader import load_csv, validate_dataframe, split_data, SAMPLE_DIR
from src.preprocessing import handle_missing_values, encode_categorical, scale_features
from src.model import train_classifier, evaluate_classifier, save_model
from src.utils import quick_eda, plot_correlation_matrix, save_plot

%matplotlib inline
sns.set_style("whitegrid")
print("✅ All libraries imported successfully!")
```

## 2. Load Sample CSV Data

```python
# Load the sample Iris CSV dataset
df = load_csv("iris_sample.csv", data_dir=str(SAMPLE_DIR))
df.head(10)
```

```python
# Validate the data
report = validate_dataframe(df, required_columns=["sepal_length", "species"])
print("Valid:", report["is_valid"])
print("Issues:", report["issues"] if report["issues"] else "None")
```

## 3. Exploratory Data Analysis

```python
# Quick EDA summary
quick_eda(df, title="Iris Sample Dataset")
```

```python
# Correlation heatmap
fig = plot_correlation_matrix(df, title="Iris Feature Correlations")
plt.show()
```

```python
# Distribution of species
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
df["species"].value_counts().plot(kind="bar", ax=axes[0], title="Species Count")
sns.boxplot(data=df, x="species", y="petal_length", ax=axes[1])
axes[1].set_title("Petal Length by Species")
plt.tight_layout()
plt.show()
```

## 4. Preprocessing

```python
# Preprocessing pipeline
df_clean = handle_missing_values(df, strategy="median")
df_encoded = encode_categorical(df_clean, columns=["species"], method="label")

print("After encoding:")
df_encoded.head()
```

## 5. Train / Test Split & Model Training

```python
# Split data
X_train, X_test, y_train, y_test = split_data(df_encoded, target_column="species")

# Train Random Forest classifier
model = train_classifier(X_train, y_train, model_name="random_forest", n_estimators=50, random_state=42)

# Evaluate
metrics = evaluate_classifier(model, X_test, y_test)
print(f"\nAccuracy: {metrics['accuracy']:.2%}")
```
